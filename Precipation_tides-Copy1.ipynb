```python
NOAATides = NOAATides.drop('Preliminary (ft)', axis=1)

for i in NOAATides.keys():
    if i == 'DateTime':
        NOAATides[i] = pd.to_datetime(NOAATides[i])
    else:
        NOAATides[i] = pd.to_numeric(NOAATides[i])
```

```python
hydrocat620['24hr move mean'] = bn.move_mean(hydrocat620['adjusted_ph temp'], window=96, min_count=1)
```

```python
tides_hydro620 = pd.merge_asof(NOAATides, hydrocat620, on="DateTime",tolerance = pd.Timedelta("30m"))
```

```python
tides_hydro620['Clock time'] = np.zeros(len(tides_hydro620['DateTime']))
for i in tides_hydro620.index:
    tides_hydro620['Clock time'][i] = tides_hydro620['DateTime'][i].strftime('%H:%M')
```

```python
trunk = {'Day':[], 'Min pH': [], 'Min pH time': [], 'Max pH': [], 'Max pH time': [],
        'Max Tide': [], 'Max Tide Time': [], 'Min Tide': [], 'Min Tide Time': []}

for day in tides_hydro620['ordinal_time'].unique():
    ds = tides_hydro620.where(tides_hydro620['ordinal_time'] == day)
    trunk['Day'].append(day)

    MINPH = ds['adjusted_ph temp'].min()
    trunk['Min pH'].append(MINPH)
    
    MAXPH = ds['adjusted_ph temp'].max()
    trunk['Max pH'].append(MAXPH)
    
    MAXTIDE = ds['Verified (ft)'].max()
    trunk['Max Tide'].append(MAXTIDE)
    
    MINTIDE = ds['Verified (ft)'].min()
    trunk['Min Tide'].append(MINTIDE)
    
    # i = Index(ds).get_loc(MINPH)
    i_min = (ds['pH total'] == MINPH).argmax()
    trunk['Min pH time'].append(ds['Clock time'][i_min])
    i_max = (ds['pH total'] == MAXPH).argmax()
    trunk['Max pH time'].append(ds['Clock time'][i_max])
    
    j_min = (ds['Verified (ft)'] == MINTIDE).argmax()
    trunk['Min Tide Time'].append(ds['Clock time'][j_min])
    j_max = (ds['Verified (ft)'] == MAXTIDE).argmax()
    trunk['Max Tide Time'].append(ds['Clock time'][j_max])
```
