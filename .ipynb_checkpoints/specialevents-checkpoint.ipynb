```python
## Mean of precipitation extreme event ( per hour ) 
MetData.set_index('DateTime', inplace = True)

# precipitation mean per hour
precmeanMet_hr = MetData['maximetPrecipitation'].resample('H').mean()
precstdMet_hr = MetData['maximetPrecipitation'].resample('H').std(ddof=1)
precNumberOfSamplesMet_hr = MetData['maximetPrecipitation'].resample('H').count()

# Calculate the standard error for DO
precstandard_errorMet = precstdMet_hr / np.sqrt(precNumberOfSamplesMet_hr)

# Calculate the upper and lower bounds for the error bars
precupper_boundMet = precmeanMet_hr + precstandard_errorMet
preclower_boundMet = precmeanMet_hr - precstandard_errorMet

## Temperature (snowed a lot in this event) 
TempmeanMet_hr = MetData['maximetTemperature'].resample('H').mean()
TempstdMet_hr = MetData['maximetTemperature'].resample('H').std(ddof=1)
TempNumberOfSamplesMet_hr = MetData['maximetTemperature'].resample('H').count()

# Calculate the standard error for DO
Tempstandard_errorMet = TempstdMet_hr / np.sqrt(TempNumberOfSamplesMet_hr)

# Calculate the upper and lower bounds for the error bars
Tempupper_boundMet = TempmeanMet_hr + Tempstandard_errorMet
Templower_boundMet = TempmeanMet_hr - Tempstandard_errorMet
```
