# Get Bottle Samples

```python
spreadsheet_key = '17FFbtUuhUS4UtxB-OjKIP2wCYJoEAmaW6VaHQPcup9U'
wks_name = 'Sheet1'


#Opening the worksheet by using Worksheet ID
spreadsheet = gc.open_by_key(spreadsheet_key)
#Selecting which sheet to pulling the data
sheet = spreadsheet.worksheet(wks_name)

#Pulling the data and transform it to the data frame
values = sheet.get_all_values()
labdata = pd.DataFrame(values[1:], columns = values[0])
labdata.head()
```

```python
df = labdata[['Sample', 'Time', 'Location', 'depth', 'TA Temp (degC)', 'TA (uM)',
       'DIC Temp (degC)', 'DIC (uM)', 'Salinity']]
df['TA (uM)'] = pd.to_numeric(df['TA (uM)'])
df['TA Temp (degC)'] = pd.to_numeric(df['TA Temp (degC)'])
df['DIC (uM)'] = pd.to_numeric(df['DIC (uM)'])
df['DIC Temp (degC)'] = pd.to_numeric(df['DIC Temp (degC)'])
df['Salinity'] = pd.to_numeric(df['Salinity'])
df["DateTime"] = pd.to_datetime(df["Time"])
df.head()
```

```python
for ind in df.index:
    # if TA or DIC at that index is Nan/empty...
    if math.isnan(df['TA (uM)'][ind]) or math.isnan(df['DIC (uM)'][ind]):
        # Drop that index
        df = df.drop(ind)
        
```

## Conversion to UTC

```python
daylight_start_date23=pd.Timestamp(datetime(2023,3,12,0,0,0))
daylight_end_date23 = pd.Timestamp(datetime(2023,11,5,0,0,0))
daylight_start_date22 = pd.Timestamp(datetime(2022,3,13,0,0,0))
daylight_end_date22 = pd.Timestamp(datetime(2022,11,6,0,0,0))
```

```python
for ind in df.index: 
    if (df['DateTime'][ind] >= daylight_start_date23) & (df['DateTime'][ind] < daylight_end_date23):
        df['DateTime'][ind]=df['DateTime'][ind]+ timedelta(hours=4)
        
    elif (df['DateTime'][ind] >=daylight_start_date22) & (df['DateTime'][ind] < daylight_end_date22):
        df['DateTime'][ind] = df['DateTime'][ind] +timedelta(hours = 4) 
        
    else:
        df['DateTime'][ind] = df['DateTime'][ind] + timedelta(hours =5)
```

## Unit conversion for bottle samples

```python
df['P']=0 #pressure

#density
df['rho_DIC'] =gsw.rho(df['Salinity'], df['DIC Temp (degC)'], df['P'])
df['rho_TA'] = gsw.rho(df['Salinity'], df['TA Temp (degC)'], df['P']) 

# converts DIC and TA 
df['DIC (umol/kg)'] = df['DIC (uM)'] / df['rho_DIC']/.001
df['TA (umol/kg)'] = df['TA (uM)'] / df['rho_TA']/.001

df.head()
```

```python
results = pyco2.sys(par1=df['TA (umol/kg)'],par2=df['DIC (umol/kg)'],par1_type=1,par2_type=2,salinity = df['Salinity'], temperature = df['DIC Temp (degC)'])
df['pH'] = results['pH']
df.head()
```

```python
df = df.dropna()
```

# Mean and SE Calculations North Data

```python
hydrocat720 = pd.DataFrame (hydrocat720) 
hydrocat720 ['DateTime'] = pd.to_datetime(hydrocat720['time'])
hydrocat720.set_index('DateTime', inplace = True)



### DO Calculations
# Calculate the least squares mean per month
least_squares_mean = hydrocat720.groupby(pd.Grouper(freq='M'))['hydrocatDissOxygen'].apply(lambda x: np.mean(np.square(x)))


# Resample the least squares mean by season
least_squares_mean_season = least_squares_mean.resample('Q-DEC').mean()

# Mean per month
mean_per_season = hydrocat720['hydrocatDissOxygen'].resample('Q-DEC').mean()

# standard error
sem_per_season = hydrocat720['hydrocatDissOxygen'].resample('Q-DEC').sem()
 
# Calculate the upper and lower bounds for the error bars
upper_bound = mean_per_season + sem_per_season
lower_bound = mean_per_season - sem_per_season


# winter_range = ((hydrocat720['time'].dt.month >= 1) & (hydrocat720['time'].dt.month <= 2)) | (df['DateTime'].dt.month == 12)
# spring_range = (hydrocat720['DateTime'].dt.month >= 3) & (hydrocat720['DateTime'].dt.month <= 5)
# summer_range = (hydrocat720['DateTime'].dt.month >= 6) & (hydrocat720['DateTime'].dt.month <= 8)
# autumn_range = (hydrocat720['DateTime'].dt.month >= 9) & (hydrocat720['DateTime'].dt.month <= 11)

# # Calculate the means for each season
# winter_mean = hydrocat720.loc[winter_range, 'hydrocatDissOxygen'].mean()
# spring_mean = hydrocat720.loc[spring_range, 'hydrocatDissOxygen'].mean()
# summer_mean = hydrocat720.loc[summer_range, 'hydrocatDissOxygen'].mean()
# autumn_mean = hydrocat720.loc[autumn_range, 'hydrocatDissOxygen'].mean()

# winter_SE = hydrocat720.loc[winter_range, 'hydrocatDissOxygen'].sem()
# spring_SE = hydrocat720.loc[spring_range, 'hydrocatDissOxygen'].sem()
# summer_SE = hydrocat720.loc[summer_range, 'hydrocatDissOxygen'].sem()
# autumn_SE = hydrocat720.loc[autumn_range, 'hydrocatDissOxygen'].sem()


```
